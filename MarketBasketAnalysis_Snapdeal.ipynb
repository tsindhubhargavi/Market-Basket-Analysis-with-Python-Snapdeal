## Task 1: Data Cleaning and Preparation

## Task 3: Customer Segmentation and Profiling

```python
df['Purchase_Frequency'] = df['Purchase_Frequency'].str.strip().str.lower()
```

```python
purchase_map = {
    'once a month': 1,
    'few times a month': 2,
    'once a week': 3,
    'multiple times a week': 4,
    'less than once a month': 0
}

df['Purchase_Frequency_Score'] = df['Purchase_Frequency'].map(purchase_map)  #Convert Purchase frequency to numeric
```

```python
df[['Purchase_Frequency','Purchase_Frequency_Score']].head()
```

```python
df['Purchase_Frequency'].unique()
```

```python
df['Purchase_Frequency_Score'].isna().sum()
```

## Creating Customer profiles

```python
def segment_customer(row):
    
    if row['Purchase_Frequency_Score'] >= 3 and row['Shopping_Satisfaction'] >= 4:
        return "Frequent Buyers"
    
    elif row['Purchase_Frequency_Score'] >= 2 and row['Shopping_Satisfaction'] >= 3:
        return "Occasional Shoppers"
    
    else:
        return "At-Risk Customers"

df['Customer_Segment'] = df.apply(segment_customer, axis=1)
```

```python
df['Customer_Segment'].value_counts()
```

## Visualizing Customer Segment

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Customer_Segment',
    data=df,
    order=df['Customer_Segment'].value_counts().index
)

plt.title("Customer Segmentation Based on Behavior")
plt.xlabel("Customer Segment")
plt.ylabel("Number of Customers")

plt.show()
```

### Analyze Segment Characteristics

```python
df.groupby('Customer_Segment')[['Shopping_Satisfaction',
                                'Customer_Reviews_Importance',
                                'Rating_Accuracy']].mean()
```

### K-Means Clustering

```python
from sklearn.cluster import KMeans

features = df[['Purchase_Frequency_Score','Shopping_Satisfaction']]

kmeans = KMeans(n_clusters=3, random_state=42)

df['Cluster'] = kmeans.fit_predict(features)
```

```python
plt.figure(figsize=(6,5))

plt.scatter(
    df['Purchase_Frequency_Score'],
    df['Shopping_Satisfaction'],
    c=df['Cluster'],
    cmap='viridis'
)

plt.xlabel("Purchase Frequency Score")
plt.ylabel("Shopping Satisfaction")
plt.title("Customer Clusters Based on Behavior")

plt.show()  #Visualizing Clusters
```
